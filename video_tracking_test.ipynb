```python
import cv2
from ultralytics import YOLO
```

```python
from cv2 import dnn_superres
```

```python

# Load the YOLOv8 model
model = YOLO('yolov8s.pt')

# Open the video file
video_path = "videos/400m_cut.mp4"
cap = cv2.VideoCapture(video_path)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:

        # mask = 
        # # Mask
        # cv2.bitwise_and(frame, frame, mask=mask)
        
        # Create an SR object
        sr = dnn_superres.DnnSuperResImpl_create()

        # Read the desired model
        path = "models/LapSRN_x8.pb"
        sr.readModel(path)

        # Set the desired model and scale to get correct pre- and post-processing
        sr.setModel("lapsrn", 8)

        # Upscale the image
        frame_upsamp = sr.upsample(frame)

        # Run YOLOv8 tracking on the frame, persisting tracks between frames
        results = model.track(frame_upsamp,
                              persist=True,
                              #tracker ='bytetrack.yaml', 
                              conf=0.08,
                              classes=0)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Tracking", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```

```python

# Load the YOLOv8 model
model = YOLO('yolov8n-pose.pt')

# Open the video file
video_path = "Nathan_20230419_short.mp4"
cap = cv2.VideoCapture(video_path)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:

        # Run YOLOv8 tracking on the frame, persisting tracks between frames
        results = model.track(frame, persist=True, tracker ='bytetrack.yaml', conf=0.05, classes=0)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("YOLOv8 Tracking", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
